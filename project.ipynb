```python
import zipfile

from PIL import Image
import pytesseract
import cv2 as cv
import numpy as np

# loading the face detection classifier
face_cascade = cv.CascadeClassifier('readonly/haarcascade_frontalface_default.xml')

# the rest is up to you!
```

```python
imagemap = {}

#iterate thru the zip file and build the image map
with zipfile.ZipFile('readonly/images.zip', 'r') as z:
    for entry in z.infolist():
        with z.open(entry) as f:
            img = Image.open(f).convert('RGB')
        imagemap[entry.filename] = { 'image' : img }
```

```python
#iterate thru each image and add text description
for key, value in imagemap.items():
    value['text'] = pytesseract.image_to_string(value['image'])
```

```python
#iterate thru the image and get the list of bounding boxes for each image
for key, value in imagemap.items():
    gray = cv.cvtColor(np.array(value['image']), cv.COLOR_BGR2GRAY)
    boxes = face_cascade.detectMultiScale(gray,1.3, 5)
    faces = []
    for x,y,w,h in boxes:
        faces.append(value['image'].crop((x, y, x+w, y+h)))
    value['faces'] = faces
    
```

```python
#convert image to thumbnail
for key, value in imagemap.items():
    for face in value['faces']:
        face.thumbnail((100, 100), Image.ANTIALIAS)
```

```python
import math
def search(keyword):
    for key,value in imagemap.items():
        if keyword in value['text']:
            if value['faces']:
                print ("Result found in file {}".format(key))
                height = math.ceil(len(value['faces'])/5.0)
                sheet = Image.new('RGB', (5*100, 100*height)) #5 images per row
                xc = yc = 0
                for face in value['faces']:
                    sheet.paste(face, (xc, yc))
                    if xc + 100 == sheet.width:
                        xc = 0
                        yc += 100
                    else:
                        xc += 100
                display(sheet)
            else:
                print ("Result found in file {}\nBut there were no faces in that file!\n".format(key))
        
```
